```python
import pandas as pd
import altair as alt
import numpy as np
np.random.seed(42)
alt.renderers.enable('notebook')
```

```python
data = pd.read_excel("Presidential Poll.xlsx", "Sheet1")
```

```python
df = pd.DataFrame(data)
```

```python
line = alt.Chart(df).mark_line().encode(
    x='Time',
    y='mean(Percentage)', 
    color='Candidate',
    #detail='index:N',
    opacity=alt.value(0.5)
).properties(width=700)

line
```
