```python
import json
from urllib.request import urlopen

with urlopen("https://data.cdc.gov/api/views/w9j2-ggv5/rows.json?accessType=DOWNLOAD") as response:
    source = response.read()
```

```python
# This converts the string from the webpage to a Json object
data = json.loads(source)
```

```python
print(json.dumps(data, indent = 2))
```

```python
age = []
year = []



for item in data["data"]:
    year.append(item[-5])
    age.append(item[-2])

diction = dict(zip(year, age))

with open("years_and_ages.json", "w") as f:
          json.dump(diction, f, indent=2)
    
```
