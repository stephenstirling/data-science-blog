```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import numpy
```

```python
data = pd.read_excel("globalterrorismdb_0718dist.xlsx")
```

```python
region_list = data['region_txt'].tolist()
country_list = data['country_txt'].tolist()
```

```python
outer_group = []
for x in region_list:
    if x not in outer_group:
        outer_group.append(x)
        
Counter(region_list)
region_amount = Counter(region_list)

outer_group_sizes = [64596, 14982, 16767]
outer_names = ['Middle East & North Africa', 'South Asia', 'South America']
```

```python
region_amount.most_common()
```

```python
inner_group = []
for x in country_list:
    if x not in inner_group:
        inner_group.append(x)
        
country_amount = Counter(country_list)
country_amount.most_common()  # sorts by most common


```

```python
inner_group_sizes = [24636, 14368, 12731, 4292, 3347, 2743, 2479, 11960, 3022, 8306, 6096, 2365]
inner_group_names = ['Iraq', 'Pakistan', 'Afghanistan', 'Turkey', 'Yemen', 'Algeria', 'Egypt', 'India', 'Sri Lanka', 
                      'Colombia', 'Peru', 'Chile']
```

```python
a, b, c, = [plt.cm.Greens, plt.cm.Reds, plt.cm.Blues]
```

```python
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(outer_group_sizes, radius=1.3, labels = outer_names,labeldistance = 1.5, colors = 
                  [a(0.8), b(0.8), c(0.8)] 
                 )
plt.setp( mypie, width=0.3, edgecolor='white')

mypie2, _ = ax.pie(inner_group_sizes, radius=1.3-0.3, labels = inner_group_names, labeldistance = 1.4, 
                   colors=[a(.7), a(.6), a(.5), a(.4), a(.3), a(.2), a(.1), b(.5), b(.4), c(.5), c(.4), c(.3)])



plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)
plt.savefig('fig1.png', dpi = 700)
plt.show()

```
